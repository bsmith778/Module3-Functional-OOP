```python
def sort012(arr):
    low = 0
    mid = 0
    high = len(arr) - 1

    while mid <= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1

    return arr

arr = [0, 1, 2, 0, 1, 2]
print(sort012(arr))
```

```python
def binary_search(arr, k):
    left = 0
    right = len(arr) - 1
    result = -1

    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == k:
            result = mid
            right = mid - 1
        elif arr[mid] < k:
            left = mid + 1
        else:
            right = mid - 1

    return result

arr = [1, 1, 1, 1, 2]
k = 1
print(binary_search(arr, k))
```

## Explanation

- The first program sorts an array of 0s, 1s, and 2s without using the built-in sort function.
- The second program uses binary search to find the first occurrence of a value in a sorted array.
